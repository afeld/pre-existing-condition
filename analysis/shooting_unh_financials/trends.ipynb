```python
import pandas as pd

file_path = "OLHCV-1m/xnas-itch-20241201-20241231.ohlcv-1m.csv.zst"
df = pd.read_csv(
    file_path,
    compression="zstd",
    parse_dates=["ts_event"],
)
df = df.sort_values("ts_event")

stock_symbol = df["symbol"].iloc[0]

df
```

```python
df["ts_event_ny"] = df["ts_event"].dt.tz_convert(tz="US/Eastern")
```

- https://news.sky.com/story/what-we-know-about-new-york-city-shooting-of-unitedhealthcare-boss-brian-thompson-13267264
- https://edition.cnn.com/us/live-news/unitedhealthcare-brian-thompson-death-12-04-24/#cm4a356lf00003b6m6nr979mg


```python
shooting_time = pd.Timestamp("2024-12-04 06:44:00", tz="US/Eastern")
shooting_label = "Brian Thompson shot"
```

```python
time_start = pd.Timestamp("2024-12-03 00:00:00", tz="US/Central")
time_end = pd.Timestamp("2024-12-05 00:00:00", tz="US/Eastern")

df_filtered = df[(df["ts_event_ny"] >= time_start) & (df["ts_event_ny"] <= time_end)]
df_filtered
```

```python
import altair as alt

base = alt.Chart(df_filtered).encode(
    x=alt.X("ts_event_ny:T", title="Time (US Eastern)"),
    y=alt.Y("close:Q", title="Close"),
)

price_line = base.mark_line().encode(
    tooltip=[
        "ts_event_ny:T",
        "open:Q",
        "high:Q",
        "low:Q",
        "close:Q",
        "volume:Q",
        "symbol:N",
    ]
)

event_df = pd.DataFrame(
    {
        "event_ts_ny": [shooting_time],
        "label": [shooting_label],
        "y": [df_filtered["close"].max()],
    }
)

event_rule = (
    alt.Chart(event_df)
    .mark_rule(color="crimson", strokeDash=[6, 4])
    .encode(x="event_ts_ny:T")
)

event_text = (
    alt.Chart(event_df)
    .mark_text(
        align="left",
        baseline="top",
        dx=6,
        dy=60,
        color="crimson",
    )
    .encode(
        x="event_ts_ny:T",
        y="y:Q",
        text="label:N",
    )
)

chart = (
    (price_line + event_rule + event_text)
    .properties(
        width=900,
        height=400,
        title=f"{stock_symbol} Close Price Over Time",
    )
    .interactive()
)

chart
```

```python
import plotly.express as px

fig = px.line(
    df_filtered,
    x="ts_event_ny",
    y="close",
    title=f"{stock_symbol} Close Price Over Time",
    labels={"ts_event_ny": "Time (US Eastern)", "close": "Close"},
)

fig.update_traces(mode="lines")
fig.add_vline(
    x=shooting_time,
    line_dash="dash",
    line_color="crimson",
    annotation_text=shooting_label,
    annotation_position="bottom left",
    annotation_font_color="crimson",
)
fig.show()
```
