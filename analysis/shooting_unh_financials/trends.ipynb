```python
import pandas as pd

file_path = "OLHCV-1m/xnas-itch-20241203-20241210.ohlcv-1m.csv.zst"
df = pd.read_csv(
    file_path,
    compression="zstd",
    parse_dates=["ts_event"],
)
df = df.sort_values("ts_event")

df
```

```python
df.dtypes
```

```python
import altair as alt

chart = (
    alt.Chart(df)
    .mark_line()
    .encode(
        x=alt.X("ts_event:T", title="Time"),
        y=alt.Y("close:Q", title="Close"),
        tooltip=[
            "ts_event:T",
            "open:Q",
            "high:Q",
            "low:Q",
            "close:Q",
            "volume:Q",
            "symbol:N",
        ],
    )
    .properties(
        width=900,
        height=400,
        title=f"{df['symbol'].iloc[0]} Close Price Over Time",
    )
    .interactive()
)

chart
```

```python
import plotly.express as px

fig = px.line(
    df,
    x="ts_event",
    y="close",
    title=f"{df['symbol'].iloc[0]} Close Price Over Time",
    labels={"ts_event": "Time", "close": "Close"},
)

fig.update_traces(mode="lines")
fig.show()
```
