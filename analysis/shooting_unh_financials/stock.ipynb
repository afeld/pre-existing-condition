```python
import pandas as pd

file_path = "OLHCV-1m/xnas-itch-20241201-20241231.ohlcv-1m.csv.zst"
df = pd.read_csv(
    file_path,
    compression="zstd",
    parse_dates=["ts_event"],
)
df = df.sort_values("ts_event")

stock_symbol = df["symbol"].iloc[0]

df
```

```python
df["ts_event_ny"] = df["ts_event"].dt.tz_convert(tz="US/Eastern")
```

```python
news_broke_start_time = pd.Timestamp("2024-12-04 07:36:00", tz="US/Eastern")
news_broke_end_time = pd.Timestamp("2024-12-04 09:00:00", tz="US/Eastern")
news_broke_label = "News of shooting breaking"
```

```python
time_start = pd.Timestamp("2024-12-03 00:00:00", tz="US/Central")
time_end = pd.Timestamp("2024-12-05 00:00:00", tz="US/Eastern")

df_filtered = df[(df["ts_event_ny"] >= time_start) & (df["ts_event_ny"] <= time_end)]
df_filtered
```

https://web.archive.org/web/20240000000000*/https://www.unitedhealthgroup.com/newsroom/2024/2024-12-3-uhg-business-outlook-investor-conference.html


```python
import plotly.express as px

fig = px.line(
    df_filtered,
    x="ts_event_ny",
    y="close",
    title=f"{stock_symbol} Stock Price Over Time",
    labels={
        "ts_event_ny": "Time (US Eastern)",
        "close": "Close Price (USD)",
    },
)

fig.update_traces(mode="lines")

for line in vlines:
    fig.add_vline(
        x=line["time"],
        line_dash="dash",
        line_color="crimson",
        annotation_text=line["label"],
        annotation_position="bottom left",
        annotation_font_color="crimson",
    )

fig.add_vrect(
    x0=news_broke_start_time,
    x1=news_broke_end_time,
    annotation_text=news_broke_label,
    annotation_position="top left",
    fillcolor="crimson",
    opacity=0.25,
    line_width=0,
)

fig.show()
```
